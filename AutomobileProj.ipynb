## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Replace '?' with NaN
df.replace("?", np.nan, inplace=True)
# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
````

````python
# Drop rows with missing categorical data (if any remain)
df.dropna(inplace=True)
````

````python
# Rename columns for clarity (optional)
df.rename(columns={'symboling': 'risk_rating'}, inplace=True)
df.head()
````

````python
# Convert columns to appropriate data types
numeric_columns = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
````

````python
# Reset index
df.reset_index(drop=True, inplace=True)
````

````python
# Final structure check
print("\nFinal Dataset Info:")
print(df.info())
````

3.Converted columns to proper numeric data types:
Columns like normalized-losses, bore, stroke, horsepower, etc., were initially strings due to "?".
Used pd.to_numeric(..., errors='coerce') to convert them.

4.Renamed confusing column:
